#### PART B: Find the range of K for stability from your Bode plots.

Since all the open-loop poles are in the left-half of the s-plane, the open-loop system is stable. Thus, the closed-loop system will be stable if the frequency response has a gain less than unity when the phase is 180°.

Looking at the Python phase plot, 180° occurs at 36.7 rad/s. Going to the magnitude plot, it's about -25.6dB at 36.7 rad/s. This means 25.6dB of gain. Converting dB to magnitude:

$$
10^{\frac{25.6}{20}} = 19.05
$$

Since magnitude plot was scaled by 5000, this value must also be scaled by 5000, so 19.05*5000 = 95,273.

So the system is stable for gain $0 < K < 95,273$.

Note: if the log magnitude plot is generated without unity scaling, so the plot doesn't start at 0dB and 1 is in the numerator instead of K, then the previous process can be done without scaling by K.

#### PART C: Evaluate gain margin, phase margin, zero dB frequency, and 180° frequency from your Bode plots for K = 10,000.

K = 10,000 is twice the amount of gain as before. This is the same as moving the magnitude curve up by $20log_{10}(2) = 6.02dB$. In *PART B*, the gain margin was found to be -25.6dB at 36.7 rad/s. Since the magnitude curve has moved up by 6.02dB, the gain margin is now 25.6dB - 6.02dB = 19.58 dB.

The phase margin is 180° - the phase at 0dB. The new mag plot starts at +6.02dB instead of 0dB, so we need to go to -6.02dB to find the new zero dB frequency. The mag curve crosses -6.02dB at 7.74 rad/s. On the phase plot, 7.74 rad/s shows 87°. Thus, the phase margin is 180° - 87° = 93°.

As previously mentioned, the new zero dB frequency is 7.74 rad/s.

The 180° frequency never changes, so this frequency is the same as what was found before in *PART B*, 36.7 rad/s.

These values can be verified in the Python plots below, which are plotted using the appropriate scaling of K = 10,000.
