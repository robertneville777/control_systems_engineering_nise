_Draw the Bode log-magnitude and phase plots for the system shown in Figure 10.10, where_

$$
G(s) = \frac{s+20}{(s+1)(s+7)(s+50)}
$$

From the poles and zeros of G(s), the break frequencies are at 1, 7, 20, and 50 rad/s. The lower limit of the frequency axis should be at least one decade less than the lowest break frequency, and the upper limit should be at least one decade greater than highest break frequency. Thus, the frequency axis limits are 0.1 and 1000 rad/s.

#### Drawing the Magnitude Response
When drawing these Bode plots by hand, using [semi-log paper](https://www.electronics-tutorials.ws/resources/printable-semi-log-graph-paper.html) and a ruler can make the process much easier.

The magnitude from 0.1 rad/s to the first break frequency at 1 rad/s is a flat line. To find the value of this flat line, substitude $s = 0$ in $G(s)$, since this is the low frequency range.

$$
G(0) = \frac{20}{(1)(7)(50)} = \frac{2}{35}
$$

Convert $\frac{2}{35}$ to dB

$$
20log_{10}(\frac{2}{35}) = -24.86 \approx -25dB
$$

Thus, the magnitude from 0.1 rad/s to 1 rad/s is a flat line at -25dB. We will now go from break frequency to break frequency, so 1 rad/s to 7 rad/s to 20 rad/s to 50 rad/s. At each break frequency, the magnitude slope will either increase by 20dB/dec if it's a zero or decrease by 20dB/dec if it's a pole.

From 1 rad/s to 7 rad/s, the slope is -20dB/dec. The slope decreases by 20dB/dec since there is a pole at 1 rad/s. My hand drawn -20dB/dec slope line intersects 7 rad/s at $\approx -43 dB$. Using (x freq, y mag) coordinates, my hand drawn line here is from (1 rad/s,-25 dB) to (7 rad/s, -43 dB).

From 7 rad/s to 20 rad/s, the slope is -40dB/dec. The slope further decreases by 20dB/dec since there is a pole at 7 rad/s. My hand drawn -40dB/dec slope line intersects 20 rad/s at $\approx -60 dB$, so my hand drawn line here is from (7 rad/s,-43 dB) to (20 rad/s, -60 dB).

From 20 rad/s to 50 rad/s, the slope is -20dB/dec. The slope **increases** by 20dB/dec since there is a zero at 20 rad/s. My hand drawn -20dB/dec slope line intersects 50 rad/s at $\approx -68 dB$, so my hand drawn line here is from (20 rad/s,-60 dB) to (50 rad/s, -68 dB).

After 50 rad/s, the slope is -40dB/dec. The slope further decreases by 20dB/dec since there is a pole at 50 rad/s. My hand drawn -40dB/dec slope line intersects 100 rad/s at $\approx -80 dB$, so my hand drawn line here is from (50 rad/s,-68 dB) to (100 rad/s, -80 dB).

My hand plotted points (1 rad/s, -25 dB), (7 rad/s, -43 dB), (20 rad/s, -60 dB), (50 rad/s, -68 dB), and (100 rad/s, -80 dB) are close to the true values in the Bode magnitude response plot below.

```python
#  Import libraries
import numpy as np
import matplotlib.pyplot as plt
import control as ct
# Make plots interactable
%matplotlib ipympl

# Generate Bode plots
numg = np.poly([-20])
deng = np.poly([-1, -7, -50])
G = ct.tf(numg, deng)

fig = plt.figure(1)
ct.bode(G, dB=True)

# Clean up workspace
del numg, deng, G, fig
```
