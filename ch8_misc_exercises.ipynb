### TryIt 8.1
Given
$$
F(S) = \frac{(s+2)(s+4)}{s(s+3)(s+6)}
$$
find $F(s)$ at the point $s = -7+j9$.

```python
import numpy as np

s = -7+9j
G = ((s+2)*(s+4))/(s*(s+3)*(s+6))
Theta = (180/np.pi)*np.angle(G)
M = abs(G)

print('Theta = ' + str(Theta))
print('M = ' + str(M))
```

### TryIt 8.2
Given a unity feedback system with the feed forward transfer function
$$
G(s) = \frac{K(s+2)}{(s^{2}+4s+13)}
$$

__a.__ Calculate the angle of $G(s)$ at the point $(-3+j0)$.\
__b.__ Determine if the point specified in __a__ is on the root locus.\
__c.__ If the point specified in __a__ is on the root locus, find the gain, K, using the lengths of the vectors.

```python
import numpy as np

s = -3+0j
G = (s+2)/(s**2+4*s+13)
Theta = (180/np.pi)*np.angle(G)
M = abs(G)
K = 1/M

print('Theta = ' + str(Theta))
print('K = ' + str(K))
```
