## Skill-Assessment Exercise 13.1

#### PROBLEM

Derive the z-transform for $f(t) = sin(\omega t)u(t)$.

#### ANSWER

I had to use the solution manual to figure this derivation. This is basically that solution with explanation to understand the steps they took in their derivation. Example 13.1 can also be followed for general guidance.

First, convert the sine function to its ideal sampled function given by Eq 13.7. Replace $t$ with $kT$ in the function $f(t)$, and substitute this $f(t)$ into Eq 13.1. The series starts at 0 instead of $-\infty$ due to the step function.

$$
f^{*}(t) = \sum_{k=-\infty}^{\infty} f(kT)\delta(t-Kt)
$$
$$
f^{*}(t) = \sum_{k=0}^{\infty} sin(\omega kT) \delta(t-Kt)
$$

Take the Laplace transform of $f^{*}(t)$ to get $F^{*}(s)$. $f^{*}(t)$ is an impulse train scaled by the sine value at each impulse, so sine can be viewed as a scaler and the main function that will be transformed is the impulse train $\delta$. The Laplace transform of $\delta(t-c)$ is $e^{cs}$. Applying this transform to $f^{*}(t)$ leads to

$$
F^{*}(s) = \sum_{k=0}^{\infty} sin(\omega kT) e^{-kTs}
$$

The goal is to convert this series into a closed form. This series is currently in the form of a general [power series](https://en.wikipedia.org/wiki/Power_series). For the power series, there are various cases in which it converges to a closed form, such as geometric, exponential, and sine series. Of these cases, the [geometric series](https://en.wikipedia.org/wiki/Power_series#Geometric_series,_exponential_function_and_sine) is the best candidate since the others use factorials. The geometric series converges to a closed form when |x| < 1:

$$
\frac{1}{1-x} = \sum_{n=0}^{\infty} x^{n} = 1 + x + x^{2} + x^{3} + ...
$$

The goal, then, is to rework the power series of $F^{*}(s)$ into a geometric series form and substitute in its closed form.

Recall that the exponential form of sine is

$$
sin(t) = \frac{e^{it}-e^{-it}}{2i}
$$

Substitute this into $F^{*}(s)$.

$$
F^{*}(s) = \sum_{k=0}^{\infty} \frac{e^{i\omega kT} - e^{-i\omega kT}}{2i} e^{-kTs}
$$

Multiply through and rework $F^{*}(s)$ so that everything is to the kth power.

$$
F^{*}(s) = \sum_{k=0}^{\infty} \frac{e^{i\omega kT -kTs} - e^{-i\omega kT-kTs}}{2i}
$$

$$
= \sum_{k=0}^{\infty} \frac{e^{(i\omega T -Ts)k} - e^{(-i\omega T-Ts)k}}{2i}
$$

Use [series multiplication](https://en.wikipedia.org/wiki/Series_(mathematics)#Scalar_multiplication) to pull out $2i$ and [series addition](https://en.wikipedia.org/wiki/Series_(mathematics)#Series_addition) to split the series to get $F^{*}(s)$ into the $x^{k}$ form.

$$
= \frac{1}{2i} \left( \sum_{k=0}^{\infty} e^{(i\omega T -Ts)k} - \sum_{k=0}^{\infty} e^{(-i\omega T-Ts)k} \right)
$$

Substitute in geometric series closed form.

$$
= \frac{1}{2i} \left( \frac{1}{1-e^{i\omega T -Ts}} - \frac{1}{1-e^{-i\omega T -Ts}} \right)
$$

After the algebraic tediousness of using a common denominator and simplifying, this will result in

$$
= \frac{e^{-Ts}}{2i} \cdot \frac{e^{-i\omega T}+e^{i\omega T}}{1} \frac{1}{1-e^{-Ts}(e^{-i\omega T}+e^{i\omega T})+e^{2Ts}}
$$

$$
= e^{-Ts} \cdot sin(\omega T) \cdot \frac{1}{1-e^{-Ts}2cos(\omega T)+e^{2Ts}}
$$

Substitute $z = e^{Ts}$. Note that this finally makes the function a function of $z$.

$$
F(z) =  \frac{z^{-1} \cdot sin(\omega T)}{1-2z^{-1}cos(\omega T)+z^{-2}}
$$

In general, it seems that the approach to finding the z-transform of a function is to convert to a geometric series to get it to closed form, simplify, and then substitute $z=e^{sT}$.
