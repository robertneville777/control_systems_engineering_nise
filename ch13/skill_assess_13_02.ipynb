## Skill-Assessment Exercise 13.2

#### PROBLEM

Find $f(kT)$ if

$$
F(z) = \frac{z(z+1)(z+2)}{(z-0.5)(z-0.7)(z-0.9)}
$$
#### ANSWER

Since the denominator of $F(z)$ is already factored, partial fraction expansion can be used to separate it into terms of $\frac{1}{1-az^{-1}}$ for which there is a common inverse z-transform pair. However, to use the partial fraction expansion method, the order of the numerator must be less than the denominator. Divide $F(z)$ by z to do this.

$$
\frac{F(z)}{z} = \frac{(z+1)(z+2)}{(z-0.5)(z-0.7)(z-0.9)}
$$

Now partial fraction expansion can be performed on the right hand side (RHS) of $F(z)$.

$$
\frac{(z+1)(z+2)}{(z-0.5)(z-0.7)(z-0.9)} = \frac{A}{z-0.5} + \frac{B}{z-0.7} + \frac{C}{z-0.9}
$$

Multiply both sides by the denominator of the left hand side (LHS).

$$
(z+1)(z+2)= A(z-0.7)(z-0.9) + B(z-0.5)(z-0.9) + C(z-0.5)(z-0.7)
$$

To solve for A, substitute $z=0.5$, which removes B and C and leaves only the A term.

$$
(0.5+1)(0.5+2)= A(0.5-0.7)(0.5-0.9)
$$

$$
\Rightarrow A = 46.875
$$

Repeat using $z=0.7$ and $z=0.9$ to find B and C. This results in $B=-114.75$ and $C=68.875$. Thus,

$$
\frac{F(z)}{z} = \frac{46.875}{z-0.5} -\frac{114.75}{z-0.7} + \frac{68.875}{z-0.9}
$$

The z-transform pairs table in the book (Table 13.1) is not good since it's missing a common pair below, which is the one we need. A quick Google search will show that this is in most z-transform tables.

$$
\frac{1}{1-az^{-1}} \leftrightarrow a^{k}u(k)
$$

Notice, however, that $\frac{F(z)}{z}$ is not quite in this form. However, if we multiply the pair above by $\frac{z}{z}$,

$$
\frac{1}{1-az^{-1}} \cdot \frac{z}{z} = \frac{z}{z-a} \leftrightarrow a^{k}u(k)
$$

then the terms of $F(z)$ are in this form. Multiply both sides of $\frac{F(z)}{z}$ by $z$.

$$
F(z) = 46.875 \cdot \frac{z}{z-0.5} -114.75 \cdot \frac{z}{z-0.7} + 68.875 \cdot \frac{z}{z-0.9}
$$

Finally, take the inverse z-transform using the z-transform pair.

$$
f(k) = 46.875(0.5)^{k} - 114.75(0.7)^{k} + 68.875(0.9)^{k}
$$
