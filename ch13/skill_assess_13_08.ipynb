#### PROBLEM

For the system of Figure 13.20 where $H(z) = 1$ and

$$
G(z) = \frac{K(z+0.5)}{(z-0.25)(z-0.75)}
$$

find the value of gain $K$ to yield the damping ratio of 0.5.

#### ANSWER
To find $K$, we just need to find where the root locus intersects the 0.5 constant damping line and see what the gain is at this intersection. Fortunately, the Python Controls Systems's library function has a `dt` parameter that can set the system to discrete time (z-plane) and display the constant damping lines.

Below, the system's root locus is plotted. Selecting the point at the intersection of the root locus and the damping ratio = 0.5 shows that the gain $K \approx 0.31$.

NOTE: Not relevant to this problem, but just for reference, the vertical marked lines are constant $\omega_{n}$.

```python
# For interacting with plots, especially selecting points on root locus plots
%matplotlib ipympl

# Import libraries
import numpy as np
import control as ct
import matplotlib.pyplot as plt

# Create transfer function
# NOTE: # With tf's dt=true, the root locus is plotted in the 
# z-plane with constant damping and wn lines displayed.
numg = np.poly([-0.5])
deng = np.poly([0.25, 0.75])
G = ct.tf(numg, deng, dt=True) 

fig = plt.figure(1) # Create and use 1st figure
ct.rlocus(G)
del numg, deng, G # Clean up workspace
```
