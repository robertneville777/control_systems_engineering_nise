## Skill-Assessment Exercise 13.5

#### PROBLEM

Determine the range of sampling interval $T$ that will make the system shown in Figure 13.16 stable.

#### ANSWER
NOTE: The first part of this problem is similar to skill-assessment 13.3.

First, the z-transform of the forward path function must be found and placed into a closed-loop. Afterwards, the range of $T$ can be found such that the poles are inside the unit circle of the z-plane.

The forward function in Figure 13.6, which will be called $G(s)$, is

$$
G(s) = \frac{1-e^{-Ts}}{s} \cdot \frac{20}{s+5}
$$

Move $s$ from the zero-order-hold factor to the factor on the right so the z-transform of the left factor can be found.

$$
G(s) = (1-e^{-Ts}) \cdot \frac{20}{s(s+5)}
$$

Take z-transform of function. Find z-transform of left factor using $z = e^{Ts}$. Leave right factor incomplete for now.

$$
G(z) = (1-z^{-1}) \cdot \mathcal{Z} \left( \frac{20}{s(s+5)} \right)
$$

Format left factor a bit nicer so it will cancel later. Multiply by $z/z$.

$$
G(z) = (1-z^{-1}) \cdot \frac{z}{z} \cdot \mathcal{Z} \left( \frac{20}{s(s+5)} \right)
$$

$$
G(z) = \frac{z-1}{z} \cdot \mathcal{Z} \left( \frac{20}{s(s+5)} \right)
$$

Focus now on completing the z-transform of the factor on the right. First, do partial fraction expansion. The steps aren't shown here. For this, see Skill-Assess 13.2.

$$
\frac{20}{s(s+5)} = 4 \left( \frac{1}{s} - \frac{1}{s+5} \right)
$$

Take z-transform.

$$
4 \left( \frac{1}{s} - \frac{1}{s+5} \right) \Leftrightarrow 4 \left( \frac{z}{z-1}-\frac{z}{z-e^{-5T}} \right)
$$

Factor out z.

$$
4z \left( \frac{1}{z-1}-\frac{1}{z-e^{-5T}} \right)
$$

Place over common denominator.

$$
4z \cdot \frac{1-e^{-5T}}{(z-1)(z-e^{-5T})}
$$

Now that the z-transform for this factor is complete, bring back the left hand side factor of $G(z)$ from before. 

$$
4z \cdot \frac{z-1}{z} \cdot \frac{1-e^{-5T}}{(z-1)(z-e^{-5T})}
$$

Simplify further by canceling terms.

$$
G(z) = \frac{4(1-e^{-5T})}{z-e^{-5T}}
$$

This $G(z)$ is the open-loop transfer function of $z$. Next, convert this to closed-loop form. Since $G(z)$ is in unity feedback, the equation for $G(z)$ in a negative feedback loop is

$$
T(z) = \frac{G(z)}{1+G(z)}
$$

Substituting $G(z)$ into the above equation results in 

$$
T(z) = \frac{4(1-e^{-5T})}{z+4-5^{-5T}}
$$

We want to determine the range of $T$ where the pole of $T(z)$ is in the unit circle. Our focus, then, is on the denominator of $T(z)$.

The pole of $T(z)$ is $z = -4+5e^{-5T}$. At $T = 0$, the pole is 1. At $T=\infty$, the pole is -4. As $T$ increases, the pole monotonically decreases from +1 to -4, i.e. the pole is moving on the real axis from +1 to -4 in the negative direction in a monotonic manner (never reversing direction). As the pole moves from +1, the pole is stable (inside the unit circle) until it passes -1, after which it leaves the unit circle and is no longer stable. Let us find the value of $T$ at which the pole crosses -1.

$$
-4+5e^{-5T} < -1
$$
$$
T < 0.1022
$$

The pole is less than +1 and greater than -1 (in the stable region of the unit circle) when $ 0 < T < 0.1022$ seconds.
