## Skill-Assessment Exercise 13.3

#### PROBLEM

Find $G(z)$ for 

$$
G(s) = \frac{8}{s+4}
$$

in cascade with a zero-order sample and hold. The sampling period is 0.25 second.

#### ANSWER

Example 13.4 is used as a guide to help solve this problem.

First, add the zero-order hold (ZOH) to $G(s)$:

$$
\frac{1-e^{-Ts}}{s} \frac{8}{s+4}
$$

Move $s$ from the ZOH denominator to the $G(s)$ denominator:

$$
(1-e^{-Ts}) \cdot \frac{8}{s(s+4)}
$$

Take the z-transform of the system. However, don't compute the z-transform of the factor on right, $G(s)/s$.

$$
(1-z^{-1})\mathcal{Z} \left( \frac{8}{s(s+4)} \right)
$$

For now, we will focus on the last term, $G(s)/s$. Perform partial fraction expansion so the z-transform can be taken on its individual terms. The partial fraction expansion steps are not shown here. For more details on how to do this, see solution for Skill Assess 13.2. The partial fraction expansion leads to:

$$
\frac{8}{s(s+4)} = 2(1-e^{-4t})
$$

Substitute $kT$ for $t$.

$$
2(1-e^{-4t}) = 2(1-e^{-4kT})
$$

Take z-transform.

$$
2(1-e^{-4kT}) = 2 \left( \frac{z}{z-1} - \frac{z}{z-e^{-4T}} \right)
$$

Combine terms in parentheses so they're over a common denominator and simplify.

$$
2 \cdot \frac{z(1-e^{-4T})}{(z-1)(z-e^{-4T})}
$$

With the z-transform of $G(s)/s$ complete, add the z-transform of the ZOH system back in.

$$
\frac{z-1}{z} \cdot 2 \frac{z(1-e^{-4T})}{(z-1)(z-e^{-4T})}
$$

Cancel factors and z's.

$$
\frac{2(1-e^{-4T})}{z-e^{-4T}}
$$

Substitute in sample time T = 0.25s and simplify.

$$
\frac{2(1-e^{-4(0.25)})}{z-e^{-4(0.25)}} = \frac{2(1-e^{-1})}{z-e^{-1}}
$$

$$
G(z) = \frac{1.264}{z-0.3679}
$$
