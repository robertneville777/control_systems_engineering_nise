## Skill-Assessment Exercise 11.1

#### PROBLEM

For a unity feedback system with a forward function 

$$
G(s) = \frac{K}{s(s+50)(s+120)}
$$

use frequency response techniques to find the value of gain $K$ to yield a closed-loop step response with a 20% overshoot.

#### ANSWER

First, select a gain $K$ value to set the low frequency response to 0dB so the Bode plots can be generated.

Since $s=0$ can't be used for the low frequency, $s = 0.1$ will be used instead. Since 0dB at this low frequency is desired, set $G(0.1) = 1$.

$$
G(0.1) = \frac{K}{0.1(0.1+50)(0.1+120)}
       = \frac{K}{0.1(50.1)(120.1)} = 1
$$

$$
\Rightarrow K = 601.701
$$

Thus, for 0dB at $\omega = 0.1$ rad/s, $K = 601.701$

Next, determine the phase margin for the given overshoot.

The phase margin equation (Eq 10.73) is given in terms of damping ratio, so overshoot % will be converted to damping ratio using Eq 4.39.

$$
\zeta = \frac{-ln(OS/100)}{\sqrt{\pi^{2}+ln(OS/100)}}
      = \frac{-ln(20/100)}{\sqrt{\pi^{2}+ln(20/100)}}
      = 0.456
$$

Putting the damp rat into the phase margin equation

$$
\Phi_{M} = tan^{-1}\frac{2\zeta}{\sqrt{-2\zeta^{2}+\sqrt{1+4\zeta^{4}}}}
         = tan^{-1}\frac{2\cdot 0.456}{\sqrt{-2\cdot 0.456^{2}+\sqrt{1+4\cdot 0.456^{4}}}}
         = 0.84 rad
         = 48.15°
$$

$K$ is the absolute value of the magnitude at the start of the phase margin. The phase margin is 48.15°, so the start of the phase margin is -180° -(-48.15) = 131.85°. The Bode plots will be used to find the frequency at 131.85° and the magnitude at this frequency. The magnitude at this frequency should be less than 0dB. $K$ is the gain needed to bring the magnitude to 0dB, e.g. the absolute value of magnitude.

```python
#  Import libraries
import numpy as np
import matplotlib.pyplot as plt
import control as ct
# Make plots interactable
%matplotlib ipympl

# Generate Bode plots for G(s)
K = 601.701 # K value to make G(s) = 0dB at w = 0.1 rad/s
numg = [K]
deng = np.poly([0, -50, -120])
G = ct.tf(numg, deng)
fig = plt.figure(1)
ct.bode(G, dB=True)

# Clean up unused variables
del K, numg, deng, G, fig
```

The phase of 131.85° occurs at $\omega$ = 27.6 rad/s. The magnitude at $\omega$ = 27.6 rad/s is -50.17 dB. Thus, the gain needs to be increased by 50.17 dB, in addition to the $K$ value of 601.701 in the beginning.

Thus the $K$ value needed is

$$
K = 601.701 \cdot 10^{50.17/20}
  = 194035
$$

This doesn't exactly match the book's answer due to the precision of picking the gain and frequency in the Bode plots. Since the magntiude is in the exponent, if it's varied even slightly, $K$ can change quite a bit.
