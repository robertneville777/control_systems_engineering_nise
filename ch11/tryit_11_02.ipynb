```python
#  Import libraries
import numpy as np
import matplotlib.pyplot as plt
import control as ct
# Make plots interactable
%matplotlib ipympl
```

```python
pos = 20    # Percent overshoot
Ts = 0.2    # Settling time
z = (-np.log(pos/100))/(np.sqrt(np.pi**2+np.log(pos/100)**2))       # Damping ratio
Pm = np.atan(2*z/(np.sqrt(-2*z**2+np.sqrt(1+4*z**4))))*(180/np.pi)  # Phase margin
Wbw = (4/(Ts*z))*np.sqrt((1-2*z**2)+np.sqrt(4*z**4-4*z**2+2))       # Closed-loop bandwidth
K = 1942000
G = ct.zpk([],[0, -50, -120], K)                                    # Construct transfer function
ct.sisotool(G,1)
```
